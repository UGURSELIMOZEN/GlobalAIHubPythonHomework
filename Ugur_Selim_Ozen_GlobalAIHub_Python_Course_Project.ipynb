#  Course Project

 Create a simple student management system :
 
 - One student must enter his/her name and surname.
 - A student who enter name and surname correctly should write "Welcome" on the screen with print function.
   The student has the 3 rights to enter his/her name and surname correctly.For more than 3 incorrect
   entries , the system shuts down and the message "Please try again later" should be printed on the screen.
 - 5 courses should be created and these courses should be kept in a list.All of these lessons should be taken from 
   the user.
 - The student can take a minimum of 3 and maximum of 5 lessons.
 - Otherwise , the mesage "You failed in class" should be returned to the student  by using return statement.
 - The student must choose one of these courses to take an exam.Add the grades from this course to a dictionary
   and keep the student's grades in a dictionary as midterm , final and project grades.
    - Example of dictionary : {"midterm":38 , "final": 66 , "project" : 89}
 - Percentages for grades
    - Midterm Exam : 30%
    - Final Exam : 50%
    - Project : 20%
 - Determine a letter grade for passing the course according to grades received.
    - For letter grades 
          - If grade >= 90 , the student will get AA.
          - If grade >= 70 and grade <90  , the student will get BB.
          - If grade >= 50 and grade <70 , the student will get CC.
          - If grade >= 30 and grade <50 , the student will get DD.
          - If grade <30 , the student will get FF.
 - If the student has received FF, the message "You failed in this course ! " should be printed on the screen.
